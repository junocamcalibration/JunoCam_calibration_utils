```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the data into memory
data = np.load('test/EZ_PJ_27_jc_000015.npy')
```

```python
# the data is structured as (H, W, channels) where channels = (UV, B, G, R, Methane) filters
data.shape
```

```python
fig, axs = plt.subplots(1, 3, dpi=150, sharex=True, sharey=True)
axs[0].imshow(data[:, :, 0], cmap='gray')
axs[1].imshow(data[:, :, 1:4][:, :, ::-1])
axs[2].imshow(data[:, :, -1], cmap='gray')
axs[0].set_xticks([])
axs[0].set_yticks([])
axs[0].set_title('UV')
axs[1].set_title('RGB')
axs[2].set_title('Methane')
plt.show()
```
